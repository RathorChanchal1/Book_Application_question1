```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df
```

```python
data = df.iloc[:,7:]
```

```python
data
```

```python
plt.xlabel('income');
plt.ylabel('value')
plt.scatter(data['median_income'], data['median_house_value'],color='red',marker='+')
```

```python
reg = linear_model.LinearRegression();
reg.fit(data[['median_income']], data.median_house_value)
```

```python
test = pd.read_csv("/content/sample_data/california_housing_test.csv")
test
```

```python
tdata = test.iloc[:,7:]
```

```python
tdata
```

```python
reg.predict(tdata[['median_income']])
```
